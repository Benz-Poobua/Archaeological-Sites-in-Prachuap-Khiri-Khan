# Prachuap Khiri Khan Map

```python
import pygmt
```

```python
grid = pygmt.datasets.load_earth_relief(resolution="01m", region=[99.07, 100.20, 10.93, 12.66])
fig = pygmt.Figure()
pygmt.config(FORMAT_GEO_MAP='ddd.x', MAP_FRAME_TYPE='plain')
fig.grdimage(
    grid=grid, 
    cmap='geo', 
    projection='M15c', 
    frame=["a", "+tPrachuap Khiri Khan"]
)
fig.grdcontour(
    annotation=100, 
    interval=100, 
    grid=grid, 
    limit=[-10, 1000]
)
fig.colorbar(frame=['x+lelevation', 'y+lm'])
fig.show()
```
